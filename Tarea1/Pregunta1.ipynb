## 1.c Datos Categ칩ricos

Ahora se deben manejar los datos no n칰mericos para poder ser ocupados en la regresi칩n.  Seg칰n columna se realizar치 un tipo distinto de transformaci칩n:

* *Gender*: Se utiliza una representaci칩n binaria, para lo cual cualquiera de los 2 g칠neros ser치 representado con un 1 칩 0 para poder ser trabajado por el algoritmo. Adem치s tendr칤a consistencia con la codificaci칩n del estado marital.
* *City_category*: Al ser una variable categ칩rica se decide centrar en 0, sin embargo se podr칤a haber usado *one_hot_encoding*.
* *Age*: Se considera ordinal no n칰merico, por lo cual cualquier codificaci칩n que represente un orden estar치 bien, tal como "1,2,3,4" para cada rango.
* *Stay_In_Current_City_Years*: Al igual que el caso anterior, se transforma a un representaci칩n ordinal n칰merica. Cambiando su 칰ltimo valor 칰nico "4+" por "4".
* *Ocupaci칩n*: Dado que a pesar de que es un dato n칰merico, cumple con ser categ칩rico por lo cual podemos codificar cada label a una posici칩n de un arreglo de 1s y 0s.

```
df.columns
```

```
mapper = {}
mapper['Gender'] = {'F':1,'M':0}
mapper['City_Category'] =  {'A':-1, 'C':0, 'B':1}
mapper['Age'] =  {'0-17':0,'18-25':1,'26-35':2,'36-45':3,'46-50':4,'51-55':5,'55+':6}
mapper['Stay_In_Current_City_Years'] = {'0':0,'1':1,'2':2,'3':3,'4+':4}
```

```
for mapp in list(mapper):
  df[mapp] = df[mapp].apply(lambda x: mapper[mapp][x])

data_pre = pd.get_dummies(df.Occupation)
col = list(df.columns)
col.remove('Occupation')
data_pre[col] = df[col]
```

```
data_pre
```

## 1.d Separaci칩n de datos

```
data_pre_train, data_pre_temp = train_test_split(data_pre, test_size= 0.25)
data_pre_val, data_pre_test = train_test_split(data_pre_temp, test_size= 0.4)
y_pre_train = data_pre_train.pop('Purchase').values
y_pre_val = data_pre_val.pop('Purchase').values
y_pre_test = data_pre_test.pop('Purchase').values
```

```
size_train = len(y_pre_train)
size_val = len(y_pre_val)
size_test = len(y_pre_test)
size_total = size_train + size_val + size_test
```

```
print("Tama침o set de entrenamiento: " + str(size_train) +"| Y su proporci칩n: " + str(size_train/size_total))
print("Tama침o set de validaci칩n: " + str(size_val) +"| Y su proporci칩n: " + str(size_val/size_total))
print("Tama침o set de pruebas/test: " + str(size_test) +"| Y su proporci칩n: " + str(size_test/size_total))
```

## 1.e Estandarizaci칩n

Ahora una buena practica antes de entrenar el modelo, es estandarizar los datos. Dado que reduce la media a 0 y la desviaci칩n estandar a 1.

```
def scale_data(data):
  scaler_x = StandardScaler()
  scaler_x.fit(data)
  return scaler_x.transform(data)
```

```
x_train = scale_data(data_pre_train)
x_val = scale_data(data_pre_val)
x_test = scale_data(data_pre_test)

y_train = scale_data(y_pre_train.reshape(-1, 1))
y_val = scale_data(y_pre_val.reshape(-1, 1))
y_test = scale_data(y_pre_test.reshape(-1, 1))
```

## 1.f Primera Regresi칩n

Obtenemos el coeficiente $R^2$ respecto a la predicci칩n, compararemos con los datos escalados y no escalados.

* Datos estandarizados:

```
def linear_reg(X,y):
  reg_scaled = LinearRegression()
  reg_scaled.fit(X,y)
  return reg_scaled
model = linear_reg(x_train, y_train)
print("Score obtenido con datos de entrenamiento: " + str(model.score(x_train,y_train)))
print("Score obtenido con datos de validaci칩n: " + str(model.score(x_val,y_val)))
print("Score obtenido con datos de prueba: " + str(model.score(x_test,y_test)))
```

* Datos no-estandarizados:

```
model = linear_reg(data_pre_train, y_pre_train)
print("Score obtenido con datos de entrenamiento: " + str(model.score(data_pre_train,y_pre_train)))
print("Score obtenido con datos de validaci칩n: " + str(model.score(data_pre_val,y_pre_val)))
print("Score obtenido con datos de prueba: " + str(model.score(data_pre_test,y_pre_test)))
```
